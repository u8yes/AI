```python
import tensorflow as tf
import numpy as np

xy = np.loadtxt('data/data-04-zoo.csv', delimiter=',', dtype=np.float32)

x_data = xy[:,0:-1]
y_data = xy[:,[-1]]

print(x_data.shape, y_data.shape)
```

```python
nb_classes = 7

y_one_hot = tf.keras.utils.to_categorical(y_data, nb_classes)
y_one_hot
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(input_dim=x_data.shape[1], units=nb_classes, activation='softmax'))
```

```python
tf.model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.SGD(learning_rate=0.1), metrics=['accuracy'])
tf.model.summary()
```

```python
tf.model.fit(x_data, y_one_hot, epochs=1000, verbose=0) # verbose=0 - 실행하는 메세지를 보여주지말라
```
