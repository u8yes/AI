```python
import tensorflow as tf
import numpy as np
```

```python
x_data = [
          [73.,80.,75.],
          [93.,88.,93.],
          [89.,91,90.],
          [96.,98.,100.],
          [73.,66.,70.]
          ]

y_data = [
          [152],
          [185],
          [180],
          [196],
          [142]
         ]
```

```python
tf.model = tf.keras.Sequential()

tf.model.add(tf.keras.layers.Dense(input_dim=3, units=1)) # input_dim=3는 feature 수를 얘기하기 때문에 차원으로 오해하지 말자.
```

```python
tf.model.compile(loss='mse',optimizer=tf.keras.optimizers.SGD(learning_rate=1e-5))
tf.model.summary()
```

```python
loss = tf.model.fit(x_data, y_data, epochs=100) # 'inf'inity
```
