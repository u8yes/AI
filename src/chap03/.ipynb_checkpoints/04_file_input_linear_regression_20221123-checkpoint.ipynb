```python
import tensorflow as tf
import numpy as np
```

```python
xy = np.loadtxt('data/data-01-test-score.csv', delimiter=',', dtype=np.float32) 
# csv의 c는 comma # delimiter=','는 구분자를 말함 # float 4byte = 32bit
x_data = xy[:,0:-1] # 슬라이싱이기 때문에 생략가능 # ,로 접근하려면 반드시 []로 감싸줘야 한다.
y_data = xy[:,[-1]]
```

```python
print(x_data, '\nx_data.shape:', x_data.shape)
print(y_data, '\ny_data.shape:', y_data.shape)
```

```python
tf.model = tf.keras.Sequential()

tf.model.add(tf.keras.layers.Dense(input_dim=3, units=1, activation='linear')) # activation 생성해달라고 변수 선언을 해서 활성화해줌.
```

```python
tf.model.summary()
```

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(learning_rate=1e-5)) 
# learning_rate가 너무 커서 loss값이 크게 나오는 것이다.
```

```python
tf.model.fit(x_data, y_data, epochs=2000)
```

```python
print(tf.model.predict([[100., 70., 98.]]))
```
