```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# 가설 설정은 H(x) = w(기울기)x + b(절편)
x_train = [1,2,3,4] # 하나의 특성은 단순회귀분석이다.
y_train = [0,-1,-2,-3] # w는 -1, b는 1

tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(input_dim=1, units=1))

sgd = tf.keras.optimizers.SGD(learning_rate=0.1)
tf.model.compile(loss='mse', optimizer='sgd')
```

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'

loss = tf.model.fit(x_train, y_train, epochs=200)

y_predict = tf.model.predict(np.array([5])) # 넘파이 이용해서 5를 예측해봄 # 결과는 -4에 가까움
y_predict
```

```python
# loss 값의 시각화
plt.plot(loss.history['loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train','Test'], loc='upper left')
plt.show()
```
